```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
pv = 120000
t = 5
m = 12
freq = t * m
r_mieszkanie = 0.05
r_lokata = 0.12
m_r_mieszkanie = r_mieszkanie/m
m_r_lokata = r_lokata/m
```

```python
periods = np.arange(1,freq+1,dtype=int)
```

```python
fv_mieszkanie_periods = fv_mieszkanie = - np.around(npf.fv(m_r_mieszkanie,periods,0,pv),2)
```

```python
fv_mieszkanie_periods[-1]
```

##### Wartość mieszkania po pięciu latach przy wzroście cen 5% rocznie wyniesie 154 003,04

```python
monthly_payemnt = - np.around(npf.ppmt(m_r_lokata,periods,freq,fv_mieszkanie_periods[-1]),2)
```

```python
np.around(monthly_payemnt.sum(),2)
```

```python
monthly_payemnt[0]
```

#### Musimy wpłacać 1 885,68 miesięcznie na lokatę 12% aby po 5 latach nasza lokata była warta 154 003,04 

```python
plt.plot(fv_mieszkanie_periods,label='wzrost ceny mieszkania: 5%')
plt.plot(monthly_payemnt.cumsum(),label='lokata 12%')
plt.title("Miesięczna wartość lokaty oraz mieszkania w ciągu 5 lat")
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('wartość')
plt.grid(True)
```
